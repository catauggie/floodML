```python
import matplotlib.pyplot as plt
import numpy as np

# Data for flood cases by month
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']
flood_cases = [23, 97, 215, 419, 612, 754, 689, 532, 318, 125, 67, 45]

# Normalize flood cases for color mapping
norm = plt.Normalize(min(flood_cases), max(flood_cases))
colors = plt.cm.viridis_r(norm(flood_cases))

# Create the figure and axes
fig, ax = plt.subplots(figsize=(12, 6))

# Create the bar histogram with gradient colors
bars = ax.bar(months, flood_cases, color=colors)

# Add color bar to indicate the scale
sm = plt.cm.ScalarMappable(cmap='viridis_r', norm=norm)
sm.set_array([])
fig.colorbar(sm, ax=ax, label='Number of Flood Cases')

# Set axis labels and title
ax.set_xticks(range(len(months)))
ax.set_xticklabels(months, rotation=45, ha='right')
ax.set_xlabel('Month')
ax.set_ylabel('Number of Flood Cases')
ax.set_title('Distribution of Flood Cases by Month in Ibadan')

# Adjust layout and show the plot
plt.tight_layout()
plt.savefig('Ibadan_floods.pdf')
plt.show()
```
